```python
#breakdown the delivery status
#we are check order late, on time or canceled
print("\n---delivery status count")
status_count= df["Delivery Status"].value_counts()
print(status_count)
```

```python
#Shipping Mode vs Delivery Status
print("\n---Shipping Mode vs Delays---")
print(pd.crosstab(df["Shipping Mode"],df["Delivery Status"]))
```

```python
#Top 5 Regions with most Late Deliveries
print("\n---Top 5 Regions with Late Delivery")
Late_orders=df[df["Delivery Status"]=="Late delivery"]
print(Late_orders["Order Region"].value_counts().head(5))
```

```python
# 3. Product Category vs Delays (Kaunsa saman late pahunch raha hai?)
print("\n--- Top 5 Product Categories with Delays ---")
print(Late_orders["Category Name"].value_counts().head(5))
```
